```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
def disks_hanoi(n):
    disks = []
    max_width = 2 * n - 1
    for i in range(n):
        width = 2 * i + 1
        x = (max_width - width) // 2
        disk = (x, n - i, width, 1)
        disks.append(disk)
    return disks
```

```python
def poles_hanoi(n):
    x = []
    posit = (2 * n - 1)
    for i in range(3):
        pole = posit * (i + 0.5) + i
        x.append(pole)
    return x
```

```python
def moves(n, left):
    if n == 0:
        return
    moves(n - 1, not left)
    if left:
        print(f'{n} to the left')
    else:
        print(f'{n} to the right')
    moves(n - 1, not left)
```

```python
def main():
    n = int(input(f'number of disks: '))
    list_disks = disks_hanoi(n)
    fig, ax = plt.subplots()
    for rect in list_disks:
        x, y, width, height = rect
        ax.add_patch(Rectangle((x, y), width, height, edgecolor='#088F8F', fill = False))
    plt.title(f'A Hanoi Tower With {n} disks')
    list_poles = poles_hanoi(n)
    for x in list_poles:
        plt.vlines(x = x, ymin = 1, ymax = n + 1, color='#8f0808')
    ax.set_xlim(-1, 4 + 3 * (2 * n - 1))
    ax.set_aspect('equal', 'datalim')
    plt.show()
    moves(n, False)
```

```python
if __name__ == '__main__':
    main()
```
