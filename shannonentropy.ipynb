```python
'''
Working with a jupyter notebook 
along with an input called shann5 that 
contains the files loaded_die.txt and 
fair_coin.txt just to test the core of the code
'''

# import sys 
import re
import numpy as np

file = open('../input/shann5/loaded_die.txt', 'r', newline = None)
_contenedor = ''

#---------------------------------------------
def isEmpty():
  global _contenedor 
  while _contenedor.strip() == '':
    line = file.readline()
    if line == '':
      return True
    _contenedor += line
 
  return False

#---------------------------------------------
def _readingE(regExp):
  global _contenedor 
  pattern = re.compile(r'^\s*' + regExp) 
  match = pattern.search(_contenedor)
  if match is None:
    raise ValueError() 
  _subS = match.group() 
  _contenedor = _contenedor[match.end():] 
  return _subS.lstrip()

#---------------------------------------------
def readInt():
  s = _readingE(r'[-+]?(0[xX][\dA-Fa-f]+|0[0-7]*|\d+)')
  return int(s, 10)

#---------------------------------------------
m = int(input('Max Int Value?:')) 
f = np.zeros(m + 1, dtype = int) 
n = 0

while not isEmpty():
  x = readInt()
  if x < m + 1 and x > 0:
    f[x] = f[x] + 1
    n = n + 1

h = 0.0
for i in range(m + 1):
  if f[i] > 0:
    p = 1.0 * f[i] / n 
    arg = -p * np.log(p) / np.log(2)
    h = h + arg

print('ShannonEntropy: ', end = '') 
print('{0:.4f}'.format(h))
```
